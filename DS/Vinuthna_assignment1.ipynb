You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````
np.random.seed(0)
X = np.random.rand(10,5)
print(X)
#max in columun
maxC = np.amax(X, axis=0)

# Get the index of elements with value 15
#result = np.where(arr == 15)
for i in range (5):
  index = np.where(X == maxC[i])
# zip the 2 arrays to get the exact coordinates
  listOfCoordinates= list(zip(index[0], index[1]))
# iterate over the list of coordinates
  for cord in listOfCoordinates:
    print('maximum in column :',cord, maxC[i] ,sep=',')
#print (index)

# min in column
minC = np.amin(X, axis = 0)
for i in range (5):
  index = np.where(X == minC[i])

  listOfCoordinates= list(zip(index[0], index[1]))

  for cord in listOfCoordinates:
    print('minimum in column :',cord, minC[i] ,sep=',')

#max in row
maxR = np.amax(X, axis=1)

# Get the index of elements with value 15
#result = np.where(arr == 15)
for i in range (10):
  index = np.where(X == maxR[i])
# zip the 2 arrays to get the exact coordinates
  listOfCoordinates= list(zip(index[0], index[1]))
# iterate over the list of coordinates
  for cord in listOfCoordinates:
    print('maximum in row :',cord, maxR[i] ,sep=',')
#print (index)

# min in row
minR = np.amin(X, axis = 1)
for i in range (10):
  index = np.where(X == minR[i])

  listOfCoordinates= list(zip(index[0], index[1]))

  for cord in listOfCoordinates:
    print('minimum in row :',cord, minR[i] ,sep=',')

#min for whole matrix
minX = np.amin(X)
index1 = np.where(X== minX)
listOfCoordinates= list(zip(index1[0], index1[1]))
for cord in listOfCoordinates:
    print('minimum in X :',cord, minX ,sep=',')

#max for whole matrix
maxX = np.amax(X)
index = np.where(X== maxX)
listOfCoordinates= list(zip(index[0], index[1]))
for cord in listOfCoordinates:
    print('maximum in X :',cord, maxX ,sep=',')
````

Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
data = pd.read_csv(file_url)
data.head()
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
df = pd.DataFrame(data, columns = ['sex','embarked','pclass'])
#data = data.drop(columns = ['sex','embarked','pclass'])
titanic_data_new=pd.get_dummies(df, columns=['sex','embarked','pclass'])
#titanic_data_new = pd.concat([data, df], axis=1, sort=False)
titanic_data_new
#df
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
am=data.age.mean()
fm=data.fare.mean()
data["age"].fillna(am, inplace = True) 
data["fare"].fillna(fm, inplace = True) 
new = pd.concat([data.age, data.fare], axis=1, sort=False)
titanic_data_new=pd.concat([titanic_data_new,new], axis=1, sort=False)
titanic_data_new
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
numeric_data = data.select_dtypes(include=[np.number])
normdata=((numeric_data-numeric_data.min())/(numeric_data.max()-numeric_data.min()))*20
normdata
````

normalized in the range 0 to 20



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
titanic_data_new=titanic_data_new.drop(['age','fare'],axis=1)
titanic_data_new=pd.concat([titanic_data_new,normdata], axis=1, sort=False)
titanic_data_new
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
plt.figure(figsize=(15,10))
plt.subplot(211)
titanic_data_new.age.plot()
plt.title('continuous plot')
plt.xlabel('ID or row number') 
plt.ylabel('age')
plt.subplot(212)
plt.title('scatter plot')
plt.xlabel('ID or row number') 
plt.ylabel('fare')
plt.scatter(titanic_data_new.index,titanic_data_new['fare'])
plt.show
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
x = np.linspace(-3*np.pi,3*np.pi,100)
y = np.sin(x)

plt.figure(figsize=(10,10))
plt.plot(x,y)

plt.xlabel('x') 
plt.ylabel('y')
plt.title(r'${y} = \sin({x})$')
plt.show
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.sin(x)
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(x,y)
plt.xlabel('x') 
plt.ylabel('y')
plt.title(r'${y} = \sin({x})$')

plt.subplot(222)
sp = np.fft.fft(y)
freq = np.fft.fftfreq(x.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.title('spectrum of sin(x)')

plt.subplot(223)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel('x') 
plt.ylabel('y')
plt.title(r'${y} = \cos({x})$')

plt.subplot(224)
sp = np.fft.fft(y)
freq = np.fft.fftfreq(x.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.title('spectrum of cos(x)')
plt.show
````
